```python
import numpy as np
def trovaspecchio(matrice):
    hp_specchio_H = []
    hp_specchio_V = []
    for riga_i in range(len(matrice)):
        if riga_i < len(matrice) - 1:
            if matrice[riga_i] == matrice[riga_i]:
                hp_specchio_H.append(riga_i)
    trasposta = np.transpose(matrice)
    for colonna_j in range(len(trasposta)):
        if colonna_j < len(trasposta) - 1:
            if trasposta[colonna_j] == trasposta[colonna_j]:
                hp_specchio_V.append(colonna_j)
    return hp_specchio_V, hp_specchio_H

    

rige_above = 0
colonne_left = 0
for pattern in patterns:
    pattern_matrix = []
    for riga in pattern:
        riga_smontata = [riga[i] for i in range(len(riga))]
        pattern_matrix.append(riga_smontata)
    
    #passo all'analisi della matrice
    hp_v, hp_h = trovaspecchio(pattern_matrix)
    print(hp_v, hp_h)
    

    
```
